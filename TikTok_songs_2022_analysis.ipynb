```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('TikTok_songs_2022.csv')


print(df.head())

```

```python
# Checking the dimensions of the dataset

df.shape
```

```python
# Summary of the dataset

df.info()
```

```python
# Checking for missing values

df.isnull().sum()
```

```python
# Descriptive statistics of numerical columns

df.describe()
```

```python
# Plotting the distribution of track popularity

plt.figure(figsize=(10, 6))
sns.histplot(df['track_pop'], bins=20, kde=True)
plt.title('Distribution of Track Popularity')
plt.xlabel('Popularity')
plt.ylabel('Count')
plt.show()
```

```python
# Visualizing the correlation matrix

plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Scatter plot of danceability vs. energy

plt.figure(figsize=(8, 6))
sns.scatterplot(x='danceability', y='energy', data=df)
plt.title('Danceability vs. Energy')
plt.show()
```

```python
#Based on the insights gained from Exploratory Data Analysis (EDA) and visualization, I can perform various analyses on the dataset
```

```python
# Grouping by artist and calculating average track popularity
avg_popularity_by_artist = df.groupby('artist_name')['track_pop'].mean().sort_values(ascending=False)

# Top 10 artists with the highest average track popularity
top_10_popular_artists = avg_popularity_by_artist.head(10)

# Bar plot of top 10 popular artists
plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_popular_artists.index, y=top_10_popular_artists.values)
plt.title('Top 10 Popular Artists')
plt.xlabel('Artist')
plt.ylabel('Average Track Popularity')
plt.xticks(rotation=45)
plt.show()
```

```python
# Top 10 tracks with the highest tempo
top_10_tempo = df.nlargest(10, 'tempo')[['track_name', 'tempo']]

# Bar plot of top 10 tracks with the highest tempo

plt.figure(figsize=(10, 6))
sns.barplot(x='tempo', y='track_name', data=top_10_tempo, palette='viridis')
plt.title('Top 10 Tracks with the Highest Tempo')
plt.xlabel('Tempo')
plt.ylabel('Track Name')
plt.show()
```

```python
# Distribution of danceability by time signature

plt.figure(figsize=(8, 6))
sns.violinplot(x='time_signature', y='danceability', data=df)
plt.title('Distribution of Danceability by Time Signature')
plt.xlabel('Time Signature')
plt.ylabel('Danceability')
plt.show()
```

```python
# Count of songs by top 10 artist
artist_counts = df['artist_name'].value_counts().head(10)

# Bar plot of song counts by artist
plt.figure(figsize=(12, 6))
sns.barplot(x=artist_counts.index, y=artist_counts.values)
plt.title('Number of Songs by Artist (Top 10)')
plt.xlabel('Artist')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Average danceability of songs

avg_danceability = df['danceability'].mean()
print("Average Danceability:", avg_danceability)
```

```python
# Calculating the proportion of acoustic tracks
acoustic_tracks = df[df['acousticness'] > 0.5]
proportion_acoustic = len(acoustic_tracks) / len(df) * 100

# Pie chart of proportion of acoustic tracks
labels = ['Acoustic Tracks', 'Non-Acoustic Tracks']
sizes = [proportion_acoustic, 100 - proportion_acoustic]
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Acoustic Tracks')
plt.axis('equal')
plt.show()
```
